```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("Countries.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.scatter(data["Longitude"],data["Latitude"])
```

```python
cluster_parameters=data.iloc[:,1:3]
cluster_parameters
```

```python
#Let us consider 3 clusters

Kmeans=KMeans(3)
```

```python
Kmeans.fit(cluster_parameters)
```

```python
Kmeans.inertia_
```

```python
predicted_clusters=Kmeans.fit_predict(cluster_parameters)
predicted_clusters
```

```python
data_copy=data.copy()
data_copy.head()
```

```python
data_copy["Pred_clusters"]=predicted_clusters
data_copy
```

```python
plt.scatter(data["Longitude"],data["Latitude"],c=data_copy["Pred_clusters"],cmap="rainbow")
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show
```

```python
#Let us consider 4 Clusters now

Kmeans_4=KMeans(4)
```

```python
Kmeans_4=Kmeans_4.fit(cluster_parameters)
wcss=Kmeans_4.inertia_
wcss
```

```python
Kmeans_4_predicted=Kmeans_4.fit_predict(cluster_parameters)
Kmeans_4_predicted
```

```python
data_copy1=data.copy()
data_copy1.head()
data_copy1["Pred_clusters"]=Kmeans_4_predicted
data_copy1
```

```python
plt.scatter(data["Longitude"],data["Latitude"],c=data_copy1["Pred_clusters"],cmap="rainbow")
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show
```
